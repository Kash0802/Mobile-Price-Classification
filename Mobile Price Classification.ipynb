```python
#drop column 

#select independent features 
X=df.drop('price_range',axis=1)
```

```python
#Normalization: when we have bell curve and data is not normailzed
from sklearn.preprocessing import Normalizer
Data_normalizer = Normalizer(norm='l1').fit(X)
Data_normalized = Data_normalizer.transform(X)
print(Data_normalized )
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2)
```

```python
model= DecisionTreeClassifier()


model2=GaussianNB()

model3 = KNeighborsClassifier(n_neighbors=2)
```

```python

#model= model.fit(X,y)
model= model.fit(X_train,y_train)

model2= model2.fit(X_train,y_train)

model3= model3.fit(X_train,y_train)


# Test model predtiction  
predictions =model.predict(X_test)

predictions2 =model2.predict(X_test)

predictions3 =model3.predict(X_test)
```

```python
#accouracy_score function returns value between 0 to 1 
from sklearn.metrics import accuracy_score 
score = accuracy_score(y_test,predictions)  
print(score)


score2 = accuracy_score(y_test,predictions2)  
print(score2)


score3 = accuracy_score(y_test,predictions3)  
print(score3)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,predictions)
c
disp = ConfusionMatrixDisplay(confusion_matrix=c, display_labels=model.classes_)
disp.plot()
plt.show()
```

```python
from sklearn.metrics import classification_report
matrix = classification_report(y_test,predictions,labels=[0,1,2,3])
print(matrix)
```
